String Formatlama

```python
name1= 'çınar'
surname = 'turan'
age1 = 36
print('my name is {1} {0}'.format(name1, surname)) #1 ve 0 index numarasıdır.

print('my name is {s} {n}'.format(n=name1, s=surname))

print("my name is {} {} and I'm {} years old.".format(name1, surname, age1))

result = 200 / 700
print('the result is {r:1.4}'.format(r=result))
#the result is'den sonra 
#1 karakter kadar boşluk/genişlik bırakılır ve .4 ile bölümün virgülden sonra 
#4 basamağının yazılması sağlanır

print(f"my name is {name1} {surname} and I'm {age} years old.")
```

Örnekler

```python
website = "http://www.google.com"
course = "python kursu: baştan sona python programlama rehberiniz (40 saat)"
```

```python
#1- 'course' karakter dizisinde kaç karakter bulunmaktadır?

length1 = len(course)

print(length1)
```

```python
#2- 'website' içinden www karakterlerini alın.

length1 = website[7:10]

print(length1)
```

```python
#3- 'website içinden com karakterlerini alın.

length1 = website[-3:]

print(length1)
```

```python
#4- 'course' içinden ilk 15 ve son 15 karakterlerini alın

length1 = course[:15]
length1 = course[-15:]

print(length1)
```

```python
#5- 'course' ifadesindeki karakterleri tersten yazdırın.

length1 = course[::-1]
print(length1)
```

```python
s = '12345' * 5
print(s[::5])
#12345 stringini 5 defa yazıp 5.indexteki elemanı yazdırır hep
#5.indeksteki eleman 1 olduğu için 1'i ard arda yazdırır.
```

```python
name, surname, age, job = 'Bora', 'Yılmaz', 32, 'mühendis'

#6- yukarıda verilen değişkenler ile ekrana aşağıdaki ifadeyi yazdırın.
#   'Benim adım Bora Yılmaz, yaşım 32 ve mesleğim mühendis

print(f"Benim adım {name} {surname}, yaşım {age} ve mesleğim {job}")
```

```python
#7- 'Hello world' ifadesindeki w harfini 'W' ile değiştirin.

s1 = 'Hello world'

d = s1[0:6] + 'W' + s1[-4:]

print(d)
```

```python
#8- 'abc' ifadesini yan yana 3 defa yazdırın.

strr= 'abc ' * 3

print(strr)
```
