```
#modules

import random as rnd
secret = rnd.randint(1,100)
check = False

for x in range(5):
  guess = int(input("please enter you guess: "))
  if guess == secret: 
    print("Congrats!!")
    check = True
    break
  elif guess < secret:
    print("please enter a greater number !!")
  else:
    print("please enter a smaller number!!")

if not check:
  print("Looseer... The number was: ", secret)
```
