## 1. Kaçış Karakteri (Escape Sequence). `\`

````python
var1 = 'Ankara\'da'
var1 = '\talan'  # \t \n \b \a
print(var1)
````

````python
var1 = "C:\tarkan\nayır\nolamaz\ben\cok\kotuyum"
print(var1)
````

````python
var1 = r"C:\tarkan\nayır\nolamaz\ben\cok\kotuyum" # raw string
print(var1)
````

````python
var1 = "C:\\tarkan\\nayır\\nolamaz\\ben\\cok\\kotuyum" # raw string
print(var1)
````

````python
print("Ali","Veli","İlkkan",sep="\n")
````

#### replace

#### split

````python
var1 = "Ali;Veli;123123\n" # comma seperated value 
var1.split(";")
````

````python
var1 = "Sadece cesurların işidir badana"
var1.split() # parametre girilmezse boşluk geçerlidir
````

````python
var1 = "Sadece cesurların işidir badana"
var1.split(maxsplit=1) # parametre girilmezse boşluk geçerlidir
````

````python
var1 = "Sadece cesurların işidir badana"
var1.rsplit(maxsplit=1) # parametre girilmezse boşluk geçerlidir
````

#### strip

````python
var1 = "    Jamiryo         "
var1.strip()
````

````python
var1 = "_____________Jami_ryo______________"
var1.strip("_")
````

````python
var1 = "___aa_ss_bb___Jami_ryo__cc____xx___"
var1.strip("_asbcx")
````

````python
var1 = "___aa_ss_bb___Jami_ryo__cc____xx___"
var1.lstrip("_asbcx") # left strip
````

````python
var1 = "___aa_ss_bb___Jami_ryo__cc____xx___"
var1.rstrip("_asbcx") # right strip
````
