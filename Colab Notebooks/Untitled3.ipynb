```
def factorial(num):
  factorial = 1
  if (num== 0 or num==1):
    print("factorial: ", factorial)
  else:
    while (num>= 1):
      factorial = factorial *num
      num-=1
      print("factorial: ", factorial)
```

```
factorial(5)
```

```
factorial(7)
```

```
def factorial(sayi):
  factorial = 1
  for i in range(1,sayi+1):
    factorial *=i
    return factorial
```
